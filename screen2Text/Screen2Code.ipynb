```python
# find the max result length:
def get_max_length(d: dict[int: str]) -> int:
    if d:
        longest = sorted([item for item in d.items()], key = lambda x: len(x[1]))[-1]
        return len(longest[1])
    return 0
```

```python
ex = ClipImg2Text()
```

```python
ex.grab()
ex.im
```

```python
start = dt.now()
ex.threads_recognize('eng', 'block')
print(f'Done in {dt.now() - start}')
```

```python
max_length = get_max_length(ex.out_texts)
print(f'{max_length}\n')

for key in ex.out_texts.keys():
    text = ex.out_texts[key]
    if max_length * .6 < len(text) < max_length * .99:
        if 'list[Fruit]' in text and '\n\n' not in text:
            print(str(key) + ': \n')
            print(text)
            print('\n' * 3)
```
